```python
from sklearn.metrics import roc_curve, roc_auc_score
import json
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the JSON file
file_path = "../results/frank/fenice_original.json"

with open(file_path, "r") as f:
    data = json.load(f)



# Extract scores and labels
scores = np.array([entry["score"] for entry in data])
# Score 값 정규화 (Min-Max Scaling)
#scores = (scores - np.min(scores)) / (np.max(scores) - np.min(scores))

labels = np.array([entry["label"] for entry in data])
# Compute ROC-AUC
roc_auc = roc_auc_score(labels, scores)

# Plot ROC Curve
fpr, tpr, _ = roc_curve(labels, scores)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, marker='.', label=f'ROC-AUC = {roc_auc:.4f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid()
plt.show()

print(f"ROC-AUC Score: {roc_auc:.4f}")
```

```python
from sklearn.metrics import precision_recall_curve, auc
import json
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the JSON file
file_path = "../results/frank/fenice_original.json"

with open(file_path, "r") as f:
    data = json.load(f)

# Extract scores and labels
scores = np.array([entry["score"] for entry in data])
labels = np.array([entry["label"] for entry in data])

# PR-AUC 계산 (모델)
precision, recall, _ = precision_recall_curve(labels, scores)
pr_auc = auc(recall, precision)

# 랜덤 분류기 PR-AUC 계산
random_scores = np.random.rand(len(labels))
random_precision, random_recall, _ = precision_recall_curve(labels, random_scores)
random_pr_auc = auc(random_recall, random_precision)

# PR Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label=f'PR-AUC (Model) = {pr_auc:.4f}')
plt.plot(random_recall, random_precision, linestyle='--', color='red', label=f'PR-AUC (Random) = {random_pr_auc:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid()
plt.show()

print(f"PR-AUC (Model): {pr_auc:.4f}")
print(f"PR-AUC (Random): {random_pr_auc:.4f}")
```
